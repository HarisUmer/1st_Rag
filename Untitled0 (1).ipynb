```python
def read_questions_answers(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    qa_pairs = []
    question = ""
    answer = ""

    for line in lines:
        stripped_line = line.strip()
        if stripped_line:
            if not question:
                question = stripped_line
            else:
                answer = stripped_line
        else:
            if question and answer:
                strr = f"{question} {answer}".lower()
                qa_pairs.append(strr)
                question = ""
                answer = ""

    # To handle the last pair if file doesn't end with a blank line
    if question and answer:
        strr = f"{question} {answer}".lower()
        qa_pairs.append(strr)

    return qa_pairs

filename = 'news1.txt'  # Replace with your filename
data_T = read_questions_answers(filename)

data_T = data_T+ read_questions_answers('news2.txt')
data_T = data_T+ read_questions_answers('news3.txt')
for pair in data_T:
      print(pair)
```

```python
from sentence_transformers import SentenceTransformer, util
model24 = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
query_embedding =[]
for pair in data_T:
  query_embedding.append(model24.encode(pair))
```

```python
import csv

# Example list of vectors (as lists)

# CSV file path
csv_file = 'VectorDB.csv'

# Write vectors to CSV file
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
     # Write header if needed
    writer.writerows(query_embedding)

print(f"Vectors have been written to {csv_file}")
```

```python
from scipy.spatial.distance import cosine

system_header = "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n\n"
user_middle = "<|eot_id|><|start_header_id|>user<|end_header_id|>\n\n"
assitant_footer = "<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n\n"
input_text = """
Tell me about David Warner?

"""
String1 = """
You are a simple  answer bot
You answer  question based on data provided
"""
encoded_input =  model24.encode(input_text)

scores = []
for sentence in data_T:
    encoded_sentence = model24.encode(sentence)
    similarity = 1 - cosine(encoded_input, encoded_sentence)
    scores.append((sentence, similarity))

# Sort scores based on similarity (highest to lowest)
scores.sort(key=lambda x: x[1], reverse=True)

# Take top 10 sentences and concatenate with input_text
top_sentences = [score[0] for score in scores[:10]]
concatenated_text =  system_header+String1+"\n" +user_middle+ " ".join(top_sentences) +" \n" + input_text+ assitant_footer



print(llm.generate(concatenated_text))
```
