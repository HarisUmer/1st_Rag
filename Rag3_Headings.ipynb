```python
import re

class RecursiveChunker:
    def __init__(self, max_chunk_size=500, max_paragraph_words=200):
        self.max_chunk_size = max_chunk_size
        self.max_paragraph_words = max_paragraph_words

    def _split_sentences(self, text):
        # Handle multi-line text within parentheses by temporarily removing newline characters
        text = re.sub(r'\(\s*\n\s*', '(', text)
        text = re.sub(r'\s*\n\s*\)', ')', text)

        # Split text into sentences using a regular expression
        sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<!\d\.)(?<!\))(?<=\.|\?|\!)\s', text)

        # Handle abbreviations by checking if they end with a period
        i = 0
        while i < len(sentences) - 1:
            if re.match(r'^[A-Z]\.$', sentences[i].strip()) or re.match(r'^[A-Z][a-z]\.$', sentences[i].strip()):
                sentences[i] = sentences[i] + ' ' + sentences[i + 1]
                del sentences[i + 1]
            else:
                i += 1

        # Combine sentences that start with '(' with the previous sentence
        i = 1
        while i < len(sentences):
            if sentences[i].startswith('('):
                sentences[i - 1] += ' ' + sentences[i]
                del sentences[i]
            else:
                i += 1

        return sentences

    def _split_paragraphs(self, text):
        # Handle multi-line text within parentheses by temporarily removing newline characters
        text = re.sub(r'\(\s*\n\s*', '(', text)
        text = re.sub(r'\s*\n\s*\)', ')', text)

        # Split text into paragraphs using double newlines
        paragraphs = text.split('\n\n')
        return paragraphs

    def _recursive_chunk_text(self, text):
        # Split text into words
        words = text.split()

        # If text is shorter than max_chunk_size, return it as a single chunk
        if len(words) <= self.max_chunk_size:
            return [text]

        # Find the best splitting point based on max_chunk_size and end of sentence
        split_point = self.max_chunk_size
        while split_point > 0 and not words[split_point - 1].endswith(('.', '!', '?')):
            split_point -= 1

        # If no good splitting point found, force split at max_chunk_size
        if split_point == 0:
            split_point = self.max_chunk_size

        # Split text into two parts and recursively chunk the remaining part
        chunk = ' '.join(words[:split_point])
        remaining_text = ' '.join(words[split_point:])
        chunks = [chunk] + self._recursive_chunk_text(remaining_text)

        return chunks

    def chunk_text(self, text):
        # Split text into paragraphs
        paragraphs = self._split_paragraphs(text)

        # Chunk each paragraph
        chunks = []
        for paragraph in paragraphs:
            # If paragraph is shorter than max_paragraph_words, add it as a single chunk
            if len(paragraph.split()) <= self.max_paragraph_words:
                chunks.append(paragraph)
            else:
                # Split paragraph into sentences
                sentences = self._split_sentences(paragraph)

                # Chunk each sentence
                sentence_chunks = []
                for sentence in sentences:
                    sentence_chunks.extend(self._recursive_chunk_text(sentence))

                # Add sentence chunks to list of chunks
                chunks.extend(sentence_chunks)

        return chunks


chunker = RecursiveChunker()
chunks = chunker.chunk_text(concatenated_string)
print(len(chunks))
```
